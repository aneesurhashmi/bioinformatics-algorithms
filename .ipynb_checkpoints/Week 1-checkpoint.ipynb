```python
# count the repitition of a pattern in a string


def pattern_count(text,pattern):
    count = 0
    for i in range(len(text[:len(text)- len(pattern)])+ 1):
        if text[i:i+len(pattern)] == pattern:
            count+=1
    return count

# pattern_count(genome,"atca")
pattern_count("CGCGATACGTTACATACATGATAGACCGCGCGCGATCATATCGCGATTATC", "CGCG")
```

```python
# find substring of size k in text/genome
def frequent_words (text, k):
    kmer_count= {}
    for i in range(len(text[:len(text)-k])):
        if text[i:i+k] not in kmer_count.keys():
            # init count
            kmer_count[text[i:i+k]] =  1
        else:
            # add count
            kmer_count[text[i:i+k]] = pattern_count(text, text[i:i+k])
            
    # return the most frequent k-mers
    max_count = max(kmer_count.values())
    most_frequent_kmer =  []
    for i in kmer_count.keys():
        if kmer_count[i] == max_count:
            most_frequent_kmer.append(i)
    return most_frequent_kmer

# find 9-mers in genome
frequent_words("CGGAGGACTCTAGGTAACGCTTATCAGGTCCATAGGACATTCA" ,3)
# frequent_words(genome1.lower(),9)
```

## Making frequent_words(text, k) better
o make FrequentWords faster, we will think about why this algorithm is slow in the first place. It slides a window of length k down Text, identifying a k mer Pattern of Text at each step. For each such k-mer, it must slide a window down the entire length of Text in order to compute PatternCount(Text, Pattern). Instead of doing all this sliding, we aspire to slide a window down Text only once. As we slide this window, we will keep track of the number of times that each k-mer Pattern has already appeared in Text, updating these numbers as we proceed.

To achieve this goal, we will first order all 4k k-mers lexicographically (i.e., according to how they would appear in the dictionary) and then convert them into the 4k different integers between 0 and 4k − 1. Given an integer k, we define the frequency array of a string Text as an array of length 4k, where the i-th element of the array holds the number of times that the i-th k-mer (in the lexicographic order) appears in Text (see figure below).



```python
# convert decimal to base n
def decimal_to_base_n(dec, base):
    base_n = []    
    while dec >= base:
        base_n.append(dec % base)
        dec = dec // base
    base_n.append(dec)
    base_n.reverse()
    base_n = [str(i) for i in base_n]
    return int("".join(base_n))

#  returns NC for given number/index
def number_to_pattern (n, k):
    n = decimal_to_base_n(n,4)
    n = str(n)
    while len(n) < k:
        n = "0"+n
    nc = ["a","c","g","t"]
#     print(f"n = {n}, return: {''.join([nc[int(i)] for i in list(n)])}")
    return "".join([nc[int(i)] for i in list(n)])

# -------------------------------------------------- #
# -------------------------------------------------- #

# returns all possible NC of given length
def get_all_kmers(k):
    return [number_to_pattern(i ,k) for i in range(4**k)]

# return index/number for given pattern 
def pattern_to_number_slow(pattern):
    nc_array = get_all_kmers(k = len(pattern))
#     print(f'nc_array: {nc_array}, pattern {pattern}')
    return nc_array.index(pattern)


def pattern_to_number(pattern):
    if pattern[0].isupper():
        pattern = pattern.lower()
    nc = ["a", "c","g","t"]
    number = [nc.index(i) for i in pattern]
    number.reverse()
    addup= 0
    for indx, value  in enumerate(number):
        addup +=value*(4**indx)
    return addup

# -------------------------------------------------- #
# -------------------------------------------------- #

# returns array with frequency of each/given NC pattern
def computing_frequencies(text,k):
    frequency_array = [0 for i in range(4**k)]
    for i in range(len(text)-k+1):
        pattern = text[i:i+k]
        pattern_index = pattern_to_number(pattern)
        frequency_array[pattern_index] = frequency_array[pattern_index]+ 1
    return frequency_array

def frequent_words_fast(text,k):
    frequency_array = computing_frequencies(text,k)
    max_count = max(frequency_array)
    frequent_patterns = [number_to_pattern(i,k) for i in range(len(frequency_array)) if frequency_array[i] == max_count]
    return frequent_patterns
```

### Finding Frequent Words by Sorting

```python
def finding_frequent_words_by_sorting(text,k):
    frequent_patterns = []
    index = []
    count = []
    for i in range(len(text)-k):
        pattern = text[i:i+k]
        # add pattern number in index 
        index.append(pattern_to_number(pattern))
        # init array count with all values = 1
        count.append(1)
    # sort index so that all repeated patterns numbers are next to each others
    sorted_index = sorted(index)
    for i in range(1,len(text)-k):
        # if any pattern is repated, they will be next to each other
        if sorted_index[i] == sorted_index[i-1]:
            count[i] = count[i-1]+1
    max_count = max(count)
#     print(count)
    for i in range(len(text)-k):
        if count[i] == max_count:
            pattern = number_to_pattern(sorted_index[i],k)
            frequent_patterns.append(pattern)
    return frequent_patterns

# finding_frequent_words_by_sorting(genome, 9)
frequent_words(genome, 9)
```

## Clump Finding  Problem

We defined a k-mer as a "clump" if it appears many times within a short interval of the genome. More formally, given integers L and t, a k-mer Pattern forms an (L, t)-clump inside a (longer) string Genome if there is an interval of Genome of length L in which this k-mer appears at least t times. (This definition assumes that the k-mer completely fits within the interval. This also does not take reverse complements into account yet.)

```python
def clump_finding(genome, k, l , t):
    if genome[0].isupper():
        genome = genome.lower()
    frequent_patterns = []
    clump = [0 for i in range(4**k)]
    for i in range(len(genome)-l):
        # windows of length l from from index i
        text = genome[i:i+l]
        frequency_array = computing_frequencies(text,k)
        for indx in range(4**k):
            if frequency_array[indx] >= t:
                clump[indx] = 1
    for i in range(4**k):
        if clump[i]==1:
            pattern = number_to_pattern(i, k)
            frequent_patterns.append(pattern)
    return frequent_patterns

clump_finding("CGGACTCGACAGATGTGAAGAACGACAATGTGAAGACTCGACACGACAGAGTGAAGAGAAGAGGAAACATTGTAA", 5, 50, 4)
```

#### Improving Clump Finding

```python
def better_clump_finding(genome,k,l,t):
    frequent_patterns = []
    clump = [0 for i in range(4**k)]
    text = genome[:l]
    frequency_array = computing_frequencies(text,k)
    for i in range(4**k):
        if frequency_array[i] >=t:
            clump[i] = 1
    for i in range(1, len(genome)-l):
        first_pattern = genome[i-1:i-1+k]
#         print(first_pattern, i)
        index = pattern_to_number(first_pattern)
        frequency_array[index] = frequency_array[index]-1
        # now for last
        last_pattern = genome[i+l-k:i+l]
        index = pattern_to_number(last_pattern)
        frequency_array[index]  = frequency_array[index]+1
        for i in range(4**k):
            if clump[i]==1:
                pattern = number_to_pattern(i, k)
                frequent_patterns.append(pattern)
    return frequent_patterns
```
